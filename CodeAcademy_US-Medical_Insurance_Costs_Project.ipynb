```python
import csv

with open('insurance.csv') as insurance:
    view_insurance = list(csv.DictReader(insurance))
    
    def get_average_age():
        age_values = []  # List to store age values
        for row in view_insurance:
            #print(row)
            age = int(row['age'])  # Convert age value to integer
            age_values.append(age)  # Add age value to the list

        # Calculate average age
        total_age = sum(age_values)
        average_age = total_age / len(age_values)

        print("Total customers:", total_age)
        print("Average age:", int(average_age))

    get_average_age()
    
    def get_total_charges():
        total = 0.0
        for row in view_insurance: 
            total += float(row['charges'])
        print('The total charges are {}'.format(round(total, 2)))
    
    get_total_charges ()
   
    def smoking_vs_nonsmoking(): 
        smokers_total = 0.0
        nonsmokers_total = 0.0
        smoker_count = 0 
        nonsmoker_count = 0
        for row in view_insurance: 
            if row['smoker'] == 'yes': 
                smokers_total += float(row['charges'])
                smoker_count += 1 
            elif row['smoker'] == 'no': 
                nonsmokers_total += float(row['charges'])
                nonsmoker_count += 1 
        print('There are {} smokers and {} non-smokers.'.format(smoker_count, nonsmoker_count))
        print('The total charges for smokers is {} compared to the charges of {} for non-smokers'.format(round(smokers_total,2),round(nonsmokers_total,2)))
       
    smoking_vs_nonsmoking()
    
    def sort_by_region():
        northeast_count = 0
        northwest_count = 0
        southeast_count = 0
        southwest_count = 0
        for row in view_insurance:
            if row["region"] == "northeast":
                northeast_count += 1
            elif row["region"] == "northwest":
                northwest_count += 1
            elif row["region"] == "southeast":
                southeast_count += 1
            elif row["region"] == "southwest":
                southwest_count += 1
        print("There are {} customers in the northeast region. {} customers in the northwest region. {} customers in the southeast region and {} customers in the southwest region".format(northeast_count, northwest_count, southeast_count, southwest_count))
      
    sort_by_region()
    
    def get_male_bmi():
        underweight = 0 
        normalweight = 0
        overweight = 0
        obese = 0
        for row in view_insurance : 
            if row['sex'] == 'male' : 
                if float(row['bmi']) < 18.5 : 
                    underweight += 1
                elif float(row['bmi']) >= 18.5 and float(row['bmi']) <= 24.9: 
                    normalweight += 1
                elif float(row['bmi']) >= 25 and float(row['bmi']) <= 29.9 : 
                    overweight += 1
                elif float(row['bmi']) >= 30 :
                    obese += 1 
        print ('{} customers are underweight \n{} customers are normalweight \n{} customers are overweight \n{} customers are obese '.format(underweight,normalweight,overweight,obese))        
        
    get_male_bmi()
    
    def get_female_bmi():
        underweight = 0 
        normalweight = 0
        overweight = 0
        obese = 0
        for row in view_insurance : 
            if row['sex'] == 'female' : 
                if float(row['bmi']) < 18.5 : 
                    underweight += 1
                elif float(row['bmi']) >= 18.5 and float(row['bmi']) <= 24.9: 
                    normalweight += 1
                elif float(row['bmi']) >= 25 and float(row['bmi']) <= 29.9 : 
                    overweight += 1
                elif float(row['bmi']) >= 30 :
                    obese += 1 
        print ('{} customers are underweight \n{} customers are normalweight \n{} customers are overweight \n{} customers are obese '.format(underweight,normalweight,overweight,obese))        
        
    get_female_bmi()
```
